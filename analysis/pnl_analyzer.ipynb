```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "../data/results/round_1_mwa_13_1.csv"
prices_df = pd.read_csv(filename, sep=";", index_col=["timestamp"], usecols=["timestamp","product","profit_and_loss"])
```

```python
prices_df
```

```python
bananas_df = prices_df[prices_df["product"] == "BANANAS"]
pearls_df = prices_df[prices_df["product"] == "PEARLS"]
cumsum_df = prices_df.groupby("timestamp").sum("profit_and_loss")
```

```python
cumsum_df.plot()
bananas_df.plot()
pearls_df.plot()
```
